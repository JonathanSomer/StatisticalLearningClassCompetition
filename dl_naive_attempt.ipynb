```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Activation, BatchNormalization
from keras import optimizers
from keras import callbacks
from keras import regularizers
```

```python
from fill_missing_values import *
from data_utils import *
# import progressbar
```

```python
def X_train_to_one_hot(X_train):
    n_samples = len(X_train)
    n_movies = len(X_train[0][1])
    
    X_one_hot = np.zeros((n_samples, n_movies, 6), dtype=int)
    
    for sample_index in range(n_samples):
        for film_index in range(n_movies):
            user_film_i_rank = X_train[sample_index, 1, film_index]
#             import pdb; pdb.set_trace()
            X_one_hot[sample_index, film_index, int(user_film_i_rank)] = 1

    return X_one_hot
```

```python
X_train, Y_train = get_X_Y_train()
X_train = fill_with_zeros(X_train)
# print(X_train[20,1])
X_train = X_train_to_one_hot(X_train)
# print(X_train[20])
```

```python
X_train, Y_train, X_validation, Y_validation = random_partition(X_train, Y_train, training_fraction = 0.9)
```

```python
from keras import backend
 
def rmse(y_true, y_pred):
    return backend.sqrt(backend.mean(backend.square(y_pred - y_true), axis=-1))

model = Sequential()
model.add(Flatten(input_shape=X_train[0].shape))
model.add(Dense(30, kernel_regularizer=regularizers.l2(0.2)))
model.add(Activation('elu'))
# model.add(Dropout(0.5, name='dropout_1'))
model.add(Dense(30, kernel_regularizer=regularizers.l2(0.2)))
model.add(Activation('elu'))
model.add(Dense(30, kernel_regularizer=regularizers.l2(0.2)))
model.add(Activation('elu'))
model.add(Dense(1))


model.compile(loss='mse', optimizer='adam', metrics=[rmse])
```

```python
callback = callbacks.ModelCheckpoint('best_model', monitor='val_loss', verbose=1, save_best_only=True, save_weights_only=False, mode='auto', period=1)

model.fit(X_train, Y_train[:,1], epochs=200, validation_data=(X_validation, Y_validation[:,1]), callbacks=[callback])

model.load_weights('best_model')
```

```python
predictions = model.predict(X_validation)
```

```python
def score_R2(y_predicted,y_actual):
    from sklearn.metrics import mean_squared_error
    from math import sqrt
    return sqrt(mean_squared_error(y_actual, y_predicted))

score_R2(predictions, Y_validation[:,1])
```
